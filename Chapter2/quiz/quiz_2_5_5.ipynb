```python
# quiz_2_5_5.ipynb cell 1

import matplotlib.pyplot as plt
from matplotlib.pyplot import grid
import numpy as np
import math

x1 = list(np.arange(-6, 3, 0.01))
y1 = [ 2**i for i in x1 ]

x2 = list(np.arange(1/2**6, 8, 0.01))
y2 = [ math.log2(i) for i in x2 ]

x3 = list(np.arange(-6, 8, 0.01))
y3 = [ i for i in x3 ]

grid()
plt.plot(x1, y1, label="y = 2^x")
plt.plot(x2, y2, label="y = math.log2(x)")
plt.plot(x3, y3, label="y = x")
plt.legend(loc='upper left')
```
