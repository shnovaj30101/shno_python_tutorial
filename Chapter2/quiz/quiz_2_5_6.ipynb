```python
# quiz_2_5_6.ipynb cell 1

import matplotlib.pyplot as plt
from matplotlib.pyplot import grid
import numpy as np
import math

x1 = list(np.arange(-9, 9, 0.01))
y1 = [ math.sin(i) for i in x1 ]

x2 = list(np.arange(-9, 9, 0.01))
y2 = [ (math.sin(i)+math.sin(3*i)/3)*2/3 for i in x2 ]

x3 = list(np.arange(-9, 9, 0.01))
y3 = [ (math.sin(i)+math.sin(3*i)/3+math.sin(5*i)/5)*1/3 for i in x2 ]

grid()
plt.plot(x1, y1, label="y = sin(x)")
plt.plot(x2, y2, label="y = (sin(x)+sin(3x)/3)*2/3")
plt.plot(x3, y3, label="y = (sin(x)+sin(3x)/3+sin(5x)/5)*1/3")
plt.legend(loc='upper left')
```
