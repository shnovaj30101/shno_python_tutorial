```python
# quiz_2_8_5.ipynb

import random
import matplotlib.pyplot as plt

def random_value_generator():
    number_of_nums = random.randint(1,10)
    nums = []
    for i in range(number_of_nums):
        nums.append(random.randint(1,500))
        
    return sum(nums)

results = []
for i in range(100000):
    results.append(random_value_generator())

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(results, range=(0, 5000), bins=100)
ax.set_xlabel('ReturnValue')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of random_value_generator 100000 Times')
plt.show()
```
