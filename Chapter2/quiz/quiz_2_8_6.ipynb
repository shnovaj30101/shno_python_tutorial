```python
# quiz_2_8_6.ipynb

import random
import matplotlib.pyplot as plt

height_results = []
weight_results = []
bmi_results = []
for i in range(100000):
    height_gene_positive_count = 0
    for j in range(100):
        height_gene_positive_count += random.randint(0,1)
        
    height = height_gene_positive_count*0.4+150
    height_results.append(height)
    
    weight_gene_positive_count = 0
    for j in range(100):
        weight_gene_positive_count += random.randint(0,1)
        
    weight = weight_gene_positive_count*0.4+50
    weight_results.append(weight)
    
    bmi = weight/height/height*10000
    bmi_results.append(bmi)
    
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(height_results, range=(150,190), bins=100, rwidth=0.8)
ax.set_xlabel('Height')
ax.set_ylabel('Count')
ax.set_title('Distribution of Height')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(weight_results, range=(50,90), bins=100, rwidth=0.8)
ax.set_xlabel('Weight')
ax.set_ylabel('Count')
ax.set_title('Distribution of Weight')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(bmi_results, range=(18,30), bins=100, rwidth=0.8)
ax.set_xlabel('BMI')
ax.set_ylabel('Count')
ax.set_title('Distribution of BMI')
plt.show()
```
