```python
# 2_12_draw_segment_graph.ipynb cell 1

import matplotlib.pyplot as plt
from matplotlib.pyplot import grid
import numpy as np
import math

x1 = list(np.arange(-5, 5, 0.01))
y1 = [ math.floor(i) for i in x1 ]

x2 = list(np.arange(-5, 5, 0.01))
y2 = [ math.ceil(i) for i in x2 ]

x3 = list(np.arange(-5, 5, 0.01))
y3 = [ round(i) for i in x2 ]

grid()
plt.plot(x1, y1, label="y = math.floor(x)")
plt.plot(x2, y2, label="y = math.ceil(x)")
plt.plot(x3, y3, label="y = round(x)")
plt.legend(loc='upper left')
```
