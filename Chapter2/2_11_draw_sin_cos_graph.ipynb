```python
# 2_11_draw_sin_cos_graph.ipynb cell 1

import matplotlib.pyplot as plt
from matplotlib.pyplot import grid
import numpy as np
import math
```

```python
# 2_11_draw_sin_cos_graph.ipynb cell 2

'''
np.arange 這個函數會生成一連串的等差數列，該函數會接收三個參數，分別是
1. 該數列的起始值，也就是下限值
2. 該數列的上限值，數列的最大值不可超過此數字
3. 等差數列的公差
''' 
x1 = list(np.arange(-math.pi, math.pi, 0.5))
x1
```

```python
# 2_11_draw_sin_cos_graph.ipynb cell 3

'''
y1 變數儲存的是一個列表，這個列表的生成方法叫做 list comprehension

簡單來說，就是從 x1 變數逐一取出每一個元素，放到變數 i，然後再將 math.sin(i) 的數值儲存到 y1 這個列表裡
'''
y1 = [ math.sin(i) for i in x1 ]
y1
```

```python
# 2_11_draw_sin_cos_graph.ipynb cell 4

'''
grid 函數可以幫繪製出來的座標圖加上座標網格線

plt.plot 必須接收兩個必要參數，分別是座標點的 x 座標列表以及 y 座標列表
然後範例中 plt.plot 的 label 參數是為該函數圖形加上一個標籤
並用 plt.legend(loc='upper left') 將該標籤顯示在左上角，來呈現該函數圖形的名稱或意義
'''

grid()
plt.plot(x1, y1, label="y = sin(x)")
plt.legend(loc='upper left')
```

```python
# 2_11_draw_sin_cos_graph.ipynb cell 5

x1 = list(np.arange(-math.pi, math.pi, 0.1))
y1 = [ math.sin(i) for i in x1 ]

x2 = list(np.arange(-math.pi, math.pi, 0.1))
y2 = [ math.cos(i) for i in x2 ]

grid()
plt.plot(x1, y1, label="y = sin(x)")
plt.plot(x2, y2, label="y = cos(x)")
plt.legend(loc='upper left')
```
